### Named Entity Recognition

```python
# Install the latest transformers, huggingface_hub, and accelerate to fix issues
!pip install -q --upgrade transformers huggingface-hub accelerate

from transformers import pipeline
import pandas as pd
```

```python
# IMPORTANT: After running the pip install above, you MUST restart the runtime
# (Runtime > Restart session) and run this cell again for the updates to take effect.
# Otherwise, you will keep seeing the "MissingSchema" or "OSError".

text = """Dear Amazon, last week I ordered an Optimus Prime action figure \
from your online store in Germany. Unfortunately, when I opened the package, \
I discovered to my horror that I had been sent an action figure of Megatron \
instead! As a lifelong enemy of the Decepticons, I hope you can understand my \
dilemma. To resolve the issue, I demand an exchange of Megatron for the \
Optimus Prime figure I ordered. Enclosed are copies of my records concerning \
this purchase. I expect to hear from you soon. Sincerely, Bumblebee."""

text2 = """Dear Nike, last week I ordered a Kevin De Bruyne Manchester City jersey \
from your online store in the United Kingdom. Unfortunately, when I opened the package, \
I discovered to my horror that I had been sent a Bruno Fernandes Manchester United shirt \
instead! As a lifelong citizen of the blue half of Manchester, I hope you can understand my \
dilemma. To resolve the issue, I demand an exchange of the United shirt for the \
City jersey I ordered. Enclosed are copies of my records concerning \
this purchase. I expect to hear from you soon. Sincerely, Erling Haaland."""

ner_tagger = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", aggregation_strategy="simple")
outputs = ner_tagger(text2)
pd.DataFrame(outputs)
```

### Question Answering

```python
reader = pipeline("question-answering")
question = "What does the customer want?"
outputs = reader(question=question, context=text2)
pd.DataFrame([outputs])
```

### Summarization

```python
summarizer = pipeline("summarization")
outputs = summarizer(text2, max_length=45, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

### Translation

```python
#translator = pipeline("translation_en_to_de",
#                      model="Helsinki-NLP/opus-mt-en-de")
#outputs = translator(text, clean_up_tokenization_spaces=True, min_length=100)
#print(outputs[0]['translation_text'])

from transformers import pipeline

# 1. Update the model to English -> Portuguese
# 'opus-mt-en-pt' covers Portuguese (including Brazilian variants)
translator = pipeline("translation",
                      model="Helsinki-NLP/opus-mt-tc-big-en-pt")

# 2. Run the translation
# Note: min_length=100 ensures a more descriptive output
outputs = translator(text2, clean_up_tokenization_spaces=True, min_length=100)

print(outputs[0]['translation_text'])

```

### Text Generation

```python
#hide
from transformers import set_seed
set_seed(42) # Set the seed to get reproducible results
```

```python
generator = pipeline("text-generation")
response = "Dear Bumblebee, I am sorry to hear that your order was mixed up."
prompt = text2 + "\n\nCustomer service response:\n" + response
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])
```
